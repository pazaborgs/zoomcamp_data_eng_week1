```python
import pandas as pd
from sqlalchemy import create_engine
from tqdm.auto import tqdm
```

```python
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/'
url = f'{prefix}/yellow_tripdata_2021-01.csv.gz'
df = pd.read_csv(url)
```

```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]
```

```python
df = pd.read_csv(
    url,
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
df.head(5)
```

```python
from sqlalchemy import create_engine
# engine = create_engine('postgresql://user:password@localhost:port/name')
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# Criando no banco de dados o SCHEMA sem dados

df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
# n=0 create tables but don't add data
```

```python
# Read CSV para 

df_iter = pd.read_csv(
    url,
    dtype =dtype,
    parse_dates=parse_dates,
    iterator = True,
    chunksize = 100000)
```

```python
first = True

for df_chunk in tqdm(df_iter):
    if first:
        df_chunk.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
        first = False
        print("Tabela criada")

    df_chunk.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")
    print("Inserted:", len(df_chunk))
```
